```python

from langchain_core.prompts import PromptTemplate

# Task 1: Create a reusable prompt template
template = "Explain {topic} in simple terms for beginners"

prompt_template = PromptTemplate(
    input_variables=["topic"],
    template=template
)

def explain_topic(topic):
    return prompt_template.format(topic=topic)
```

```python
print(explain_topic("Artificial Intelligence"))
```

```python

from langchain_core.prompts import PromptTemplate

# Task 2: Multi-input template
template = "Explain {topic} for {audience} in a {tone} tone"

# Create PromptTemplate with multiple inputs
multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Function to generate prompt
def generate_multi_prompt(topic, audience, tone):
    return multi_prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
print(generate_multi_prompt("AI", "beginners", "friendly"))
print(generate_multi_prompt("Python", "kids", "fun"))
print(generate_multi_prompt("Deep Learning", "engineers", "technical"))
```

```python

from langchain_core.prompts import PromptTemplate

# Task 3: Create different prompt styles

# 1. Teaching Style
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# 2. Interview Style
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# 3. Storytelling Style
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)
```

```python
topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```python

from langchain_core.prompts import ChatPromptTemplate

# Task 4: Function to create chat prompt based on role
def create_chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher who explains concepts clearly."

    elif role == "interviewer":
        system_message = "You are an interviewer who asks insightful questions."

    elif role == "motivator":
        system_message = "You are a motivational speaker who inspires learners."

    else:
        system_message = "You are a helpful assistant."

    # Create chat template
    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Tell me about {topic}")
    ])

    return chat_template.format(topic=topic)
```

```python
result = create_chat_prompt("teacher", "Neural Networks")
print(result)
```

```python
# Task 5: Input validation function

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Validate audience
    if audience not in valid_audience:
        print("Invalid audience! Setting default = 'beginner'")
        audience = "beginner"

    # Validate tone
    if tone not in valid_tone:
        print("Invalid tone! Setting default = 'formal'")
        tone = "formal"

    return audience, tone
```

```python
print(validate_inputs("beginner", "fun"))        # valid
print(validate_inputs("kids", "cool"))          # invalid
```

```python

from langchain_core.prompts import PromptTemplate

# Task 6: Prompt Generator Function

def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Select template based on style

    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"

    elif style == "interview":
        template = "Ask interview questions about {topic} for a {audience} audience in a {tone} tone"

    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"

    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    # Step 3: Create PromptTemplate
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    # Step 4: Generate final prompt
    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
